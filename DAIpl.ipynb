## Total Matches Played by a team

```python
teams_matches = pd.concat([df['Team1'],df['Team2']]).value_counts()
teams_matches = pd.DataFrame(teams_matches).reset_index()
teams_matches.rename(columns={'index':'Teams','count':'Total_Matches'},inplace=True)
```

```python
teams_matches
```

## Visualization part: Matplotlib

```python
def get_color(name, number):
    pal = list(sns.color_palette(palette=name, n_colors=number).as_hex())
    return pal
```

```python
pal_vi = get_color('viridis_r', len(teams_matches))
pal_vi
```

```python
# Set figure size
plt.figure(figsize=(30, 10))

# Create bar plot
teams_matches1 = teams_matches.head(10)
bars = plt.bar(
    teams_matches1['Teams'],
    teams_matches1['Total_Matches'],
    color=pal_vi[:len(teams_matches1)]  # Use your color palette
)


plt.xlabel('Teams')
plt.ylabel('Total Matches')
for bar in bars:
    height = bar.get_height()
    plt.text(
        bar.get_x() + bar.get_width() / 2,
        height,
        f'{height}',
        ha='center',
        va='bottom',
        fontsize=10
    )

# Customize plot appearance
plt.xlabel('Teams')
plt.ylabel('Total Matches')
plt.xticks(rotation=45)
plt.gca().set_facecolor('white')   # Plot background
plt.gcf().set_facecolor('white')   # Figure background

# Remove top and right spines for a cleaner look
plt.gca().spines['top'].set_visible(False)
plt.gca().spines['right'].set_visible(False)

# Adjust layout
plt.tight_layout()
plt.show()
```

## Is win depend on winning toss?

```python
total = (df.TossWinner==df.WinningTeam).value_counts().sum()
```
